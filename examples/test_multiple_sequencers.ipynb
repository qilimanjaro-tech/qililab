```python
from qililab import Domain, Gaussian, IQPair, QProgram, Square
from qililab.qprogram import QbloxCompiler
```

```python
qp = QProgram()
frequency = qp.variable(label="frequency", domain=Domain.Frequency)
with qp.average(shots=1000):
    with qp.for_loop(variable=frequency, start=10e6, stop=100e6, step=10e6):
        qp.set_frequency(bus="drive_bus", frequency=frequency)
        qp.play(bus="drive_bus", waveform=IQPair(I=Square(1.0, 500), Q=Square(1.0, 500)))
        qp.sync()
        qp.measure(
            bus="readout_bus",
            waveform=IQPair(I=Gaussian(1.0, 200, 4.5), Q=Square(1.0, 200)),
            weights=IQPair(I=Square(1.0, 2000), Q=Square(1.0, 2000)),
        )
```

```python
compiler = QbloxCompiler()
output = compiler.compile(qprogram=qp)
```
